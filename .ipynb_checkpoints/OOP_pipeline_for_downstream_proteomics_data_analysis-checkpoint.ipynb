## Preprocessing Class

```python
class Preprocessing:
    
    """
    This class encapsulates preprocessing steps
    """
    
    def __init__(self):
        
        print()
        print('Preprocessing object has been created')
        print()
        
    def change_datatypes(self, data):
        
        """
        change datatypes
        """
        
        self.data = data.copy()
        for i in self.data.columns.unique():
            if i in ['Protein.Group', 'Protein.Ids', 'Protein.Names', 'Genes','First.Protein.Description']:
                self.data[i] = self.data[i].astype('string')
            else:
                self.data[i] = self.data[i].astype('float') 
        return self.data
    
    def log2_transform(self, data):
        
        """
        log2 transfrom float data
        """

        for i in self.data.select_dtypes('float').columns.unique():
            self.data[i] = np.log2(self.data[i])
        return self.data
        
    def filter_for_valid_vals_in_one_exp_group(self, data):
        
        """
        Filter for valid values in at least one experimental group
        param data: diann output table
        """
        
        columns_name = self.data.select_dtypes('float').columns.unique()
        
        indices = []
        
        for col in columns_name:
            _ = self.data[self.data[col].isna().sum(axis=1)==0].index.to_list()
            indices += _
        indices = list(set(indices))
        self.data = self.data[self.data.index.isin(indices)].reset_index(drop = True)
        
        return self.data
    
    def impute_genes_with_protein_ids(self, data):
        
        """
        Imputing missing values in Genes column with the protein IDs
        """
        imputation_values = [i.split(';')[0] for i in self.data[self.data['Genes'].isna()]['Protein.Ids'].values]
        
        for idx, value in zip(self.data[self.data['Genes'].isna()]['Protein.Ids'].index, imputation_values):
            self.data.loc[idx, ['Genes']] = value
        return self.data
    
    def impute_based_on_gaussian(self, data):
         
        """
        This function imputes missing values in a dataset using a Gaussian distribution.
        The missing values are imputed by random sampling values from a Gaussian distribution with a mean
        of 3 standard deviations below the computed mean and a width of 0.3 times the computed standard deviation.
        The function returns a copy of the imputed dataset with the missing values replaced.
        """
        # select only float data
        float_data = self.data.select_dtypes('float')
        na_data = float_data[float_data.isna().any(axis=1)]

        # loop through na_data rows and generate imputed data
        imp_array = []

        for i in na_data.index: 
            na_count = na_data.loc[i].isna().sum()

            mean = na_data.loc[i].mean(skipna=True)
            std = na_data.loc[i].std(skipna=True)

            np.random.seed(i)
            imp_array.append(list(np.random.normal(loc=mean-3*std, scale=0.3*std, size=na_count)))

        imp_values_list = np.array([imp_value for innerlist in imp_array for imp_value in innerlist])

        copied_data = float_data.copy()
        columns = float_data.columns
        copied_data.columns = np.arange(copied_data.shape[1])
        
        # impute the data
        stacked_na_data = copied_data.stack(dropna=False)
        na_index = stacked_na_data[stacked_na_data.isna()].index
        stacked_na_data.loc[na_index] = imp_values_list
        imp_data = stacked_na_data.unstack()
        imp_data.columns = columns
        imp_data = imp_data.merge(data.select_dtypes('string'), left_index=True, right_index=True)
        self.data = imp_data.copy()
        
        return self.data
```

## Preprocessor Class

```python
class Preprocessor:
    
    """
    This class applies preprocessing steps to DIANN data
    """
    
    def __init__(self):
        self.data = None
        self._preprocessing = Preprocessing()
    
    def _process(self, data):
        
        self.data = data
        
        # change the datatypes
        self.data = self._preprocessing.change_datatypes(self.data)
        
        # log2 transform data
        self.data = self._preprocessing.log2_transform(self.data)
        
        # filter for valid values in at least one experimental group
        self.data = self._preprocessing.filter_for_valid_vals_in_one_exp_group(self.data)
        
        # impute genes data
        self.data = self._preprocessing.impute_genes_with_protein_ids(self.data)
        
        # impute data based on normal distribution
        self.data = self._preprocessing.impute_based_on_gaussian(self.data)
        
        return self.data
```

# Visualizer Class

```python
class Visualizer:
    
    def __init__(self):
        pass
    
    def pca_fig(self, figsize = (6,6), savefig=False):
    
        '''Generate a PCA plot'''

        # prepare the dataframe for PCA
        df = self.data.select_dtypes(include =['float64']).T.copy()

        # prepare features variable (Genes)
        features = list(np.arange(df.shape[1]))
        df.columns = features

        # prepare target column (sample types)
        df['target'] = df.index
        df = df.reset_index(drop= True)

        # Separating out the features
        x = df.loc[:, features].values

        # Separating out the target
        y = df.loc[:,['target']].values

        # Standardizing the features
        x = StandardScaler().fit_transform(x)
        pca = PCA(n_components=2)
        principalComponents = pca.fit_transform(x)
        principalDf = pd.DataFrame(data = principalComponents
                                  ,columns = ['principal component 1', 'principal component 2'])

        finalDf = pd.concat([principalDf, df[['target']]], axis = 1)

        fig = plt.figure(figsize = figsize)

        # prepare scatterplot 
        sns.scatterplot(data = finalDf, x = 'principal component 1', y = 'principal component 2', hue = 'target')
        pc1 = "%.2f" % (pca.explained_variance_ratio_[0]*100)
        pc2 = "%.2f" % (pca.explained_variance_ratio_[1] *100)
        plt.xlabel(f'PC1 {pc1}%')
        plt.ylabel(f'PC2 {pc2}%')
        plt.legend(bbox_to_anchor=(1.6, 1), loc='upper right', borderaxespad=0)
        sns.despine()

        if savefig == True:
            plt.savefig('pca.pdf', transparent=True, bbox_inches=True)

        plt.show()
        
    def volcano_eda(self, n_rows, n_cols, gene_list=[], figsize = (8,8), upper_fc_cutoff = 1, lower_fc_cutoff = 1,
                 annot = None, gene_column = 'Genes', q_value=True):
    
        '''
        Plot a volcano plot 
        gene_list = annotate the datapoints with Gene names in gene list
        upper_fc_cutoff = 
        annot = use conditions to make annotations of Gene names to datapoints ("fc_cutoff", "None")
        upper_fc_cutoff = use upper fc cutoff, draws a line at cutoff and annotates datapoints that cross fc cutoff
        lower_fc_cutoff = use lower fc cutoff, draws a line at cutoff and annotates datapoints that cross fc cutoff
        qvalue = True when qvalue is used for significance cutoff
        '''
        fc_data = self.fc_data
        pv_data = self.pv_data
        qv_data = self.qv_data
        
        col_names = fc_data.select_dtypes('float').columns
        
        fig, ax = plt.subplots(n_rows, n_cols)
        fig.tight_layout(pad=0.5)
        fig.set_figheight(4)
        fig.set_figwidth(14)
        for i, axes in zip(col_names, ax.flat):

            # plot scatterplot with 5% FDR coloring
            if (qv_data[i] < 0.05).sum(axis=0) > 0:
                sns.scatterplot(x = fc_data[i], y = pv_data[i], color = 'lightgrey', alpha = 0.5, ax=axes)
                sns.scatterplot(x = fc_data[qv_data.select_dtypes('float') < 0.05][i],
                                y = pv_data[qv_data.select_dtypes('float') < 0.05][i], color = 'lightcoral', ax=axes)
                sns.scatterplot(x = fc_data[(qv_data.select_dtypes('float') < 0.05) & (fc_data.select_dtypes('float') < 0)][i],
                                y = pv_data[(qv_data.select_dtypes('float') < 0.05) & (fc_data.select_dtypes('float') < 0)][i], color = 'cornflowerblue', ax=axes)
                # set q-value threshold
                idx_threshold = qv_data[qv_data[i] < 0.05][i].sort_values(ascending=False).index[0]
                threshold = pv_data.loc[idx_threshold][i]
                axes.axhline(threshold,ls = '--', color = 'lightgrey')

            else:
                plt.figure(figsize = figsize)
                sns.scatterplot(x = fc_data[i], y = pv_data[i], color = 'lightgrey', alpha = 0.5, ax=axes)
                sns.scatterplot(x = fc_data[pv_data.select_dtypes('float') > 1.3][i],
                                y = pv_data[pv_data.select_dtypes('float') > 1.3][i], color = sns.color_palette("deep", 20)[1], ax=axes)
                sns.scatterplot(x = fc_data[(pv_data.select_dtypes('float') > 1.3) & (fc_data.select_dtypes('float') < 0)][i],
                                y = pv_data[(pv_data.select_dtypes('float') > 1.3) & (fc_data.select_dtypes('float') < 0)][i], color = sns.color_palette("deep", 20)[2], ax=axes)
                # set p-value threshold 
                idx_threshold = pv_data[pv_data[i] > 1.3][i].sort_values(ascending=True).index[0]
                threshold = pv_data.loc[idx_threshold][i]
                axes.axhline(threshold,ls = '--', color = 'lightgrey')

            # Gene annotation
            if annot == 'fc_cutoff' and qvalue:
                indices = fc_data[(fc_data[i] < lower_fc_cutoff) | (fc_data[i] > upper_fc_cutoff) & (qv_data[i] <0.05)].index
                axes.axvline(lower_fc_cutoff, ls = '--', color = 'lightgrey')
                axes.axvline(upper_fc_cutoff, ls = '--', color = 'lightgrey')

            elif annot == 'fc_cutoff' and qvalue == False:
                indices = fc_data[(fc_data[i] < lower_fc_cutoff) | (fc_data[i] > upper_fc_cutoff) & (pv_data[i] >1.3)].index
                axes.axvline(lower_fc_cutoff, ls = '--', color = 'lightgrey')
                axes.axvline(upper_fc_cutoff, ls = '--', color = 'lightgrey')

            elif annot == None: 
                indices = fc_data[fc_data[gene_column].isin(gene_list)].index

            sns.scatterplot(x = fc_data.loc[indices][i],
                            y = pv_data.loc[indices][i], color = 'lightblue', ax=axes)
            sns.despine()

            # annoation 
            texts = [axes.text(fc_data[i][idx],pv_data[i][idx], fc_data[gene_column][idx], ha='center', va='center') for idx in indices]
            adjust_text(texts, arrowprops = dict(arrowstyle = '-', color = 'black'), ax=axes)
            
            axes.set(xlabel=f'log2 fold change ({i})', ylabel=f'-log10 p-value ({i})')
            
    def volcano_plotly(self, n_rows, n_cols, gene_column = 'Genes', width=800, height=400, protein_description_column = 'First.Protein.Description'):
        
        fc_data_plotly = self.fc_data.copy()
        pv_data_plotly = self.pv_data.copy()
        qv_data_plotly = self.qv_data.copy()

        str_col_names = ['Protein.Group', 'Protein.Ids', 'Protein.Names', 'Genes','First.Protein.Description']
        
        fc_data_plotly.columns = str_col_names + [i + '(fc)' for i in fc_data_plotly.select_dtypes('float').columns] 
        pv_data_plotly.columns = str_col_names + [i + '(pv)' for i in pv_data_plotly.select_dtypes('float').columns] 
        qv_data_plotly.columns = str_col_names + [i + '(qv)' for i in qv_data_plotly.select_dtypes('float').columns] 

        # merge all data
        fc_dataframes = [fc_data_plotly, pv_data_plotly, qv_data_plotly]
        merged_data = reduce(lambda  left,right: pd.merge(left,right,on=str_col_names, how='inner'), fc_dataframes)

        fc_columns = fc_data_plotly.select_dtypes('float').columns
        pv_columns = pv_data_plotly.select_dtypes('float').columns
        qv_columns = qv_data_plotly.select_dtypes('float').columns

        fig = make_subplots(rows=n_rows, cols=n_cols)
        row_col_number = [(i, j) for i in range(1, n_rows+1) for j in range(1, n_cols+1)]

        for fc, pv, qv, axes in zip(fc_columns, pv_columns, qv_columns, row_col_number):

            # determine the axes in subplot
            row, col = axes

            # prepare data for plotting
            plot_data =  merged_data[[fc,pv,qv]]
            plot_data = plot_data.copy()
#             plot_data['FDR'] = plot_data[qv] < 0.05
#             color_discrete_map = {True: 'red', False: 'blue'}
            plot_data['Gene & Description'] = [gene+', '+description for gene, description in zip(merged_data['Genes'], merged_data['First.Protein.Description'])]
#             plot_data = plot_data.merge(merged_data[[gene_column, protein_description_column,]], left_index = True, right_index = True)

            fig.add_trace(go.Scatter(x = plot_data[fc], y= plot_data[pv], mode='markers', text=plot_data['Gene & Description']), row=row, col=col)
            fig.update_xaxes(title_text=fc, row=row, col=col)
            fig.update_yaxes(title_text=pv, row=row, col=col)
        fig.update_layout(height=500, width=1200,)
        fig.show()
```

# Stats class

```python
class Statistics:
    
    def __init__(self):
        pass
    
    def unpaired_two_tailed_ttest(self, comparisons):
        
        '''Unpaired two tailed student's t-test with BH pvalue adjustment and fold change calculation'''
    
        # create fc_data, pv_data, qv_data
        fc_data = pd.DataFrame()
        pv_data = pd.DataFrame()
        qv_data = pd.DataFrame()

        for a, b in comparisons:

            a_data = self.data[a].to_numpy()
            b_data = self.data[b].to_numpy()

            # T test for each row pair.
            pvalues = stats.ttest_ind(a=a_data, b=b_data, axis=1)[1]

            pv_data[f'{a}/{b}'] = - np.log10(pvalues)

            # Mean difference for each row.
            fc_data[f'{a}/{b}'] = np.mean(a_data, axis=1) - np.mean(b_data, axis=1)

            # bh adjusted pvalue 
            qv_data[f'{a}/{b}'] = fdrcorrection(pvalues)[1]

        fc_data = self.data.select_dtypes('string').merge(fc_data, left_index = True, right_index = True)
        pv_data = self.data.select_dtypes('string').merge(pv_data, left_index = True, right_index = True)
        qv_data = self.data.select_dtypes('string').merge(qv_data, left_index = True, right_index = True)
        
        self.fc_data = fc_data
        self.pv_data = pv_data
        self.qv_data = qv_data
```

## DIANN Data Class

```python
class DiannDataObjectOriented(Statistics, Visualizer):
    
    """
    param data: diann output data
    """
    
    def __init__(self):
        # properties
#         self.fc_data = None
#         self.pv_data = None
#         self.qv_data = None
#         self.data = None
#         self.rows = None
#         self.cols = None
#         self.col_names = None
#         self.missing_values = None
        
        self._Preprocessor = Preprocessor()
        
        print()
        print('DiannDataObjectOriented Object has been created.')
        print()
    
    def add_data_and_change_col_names(self, data, col_names):
            
        """
        change the colums names of DIANN output tables
        replicates will have same name
        param data: DIANN output table
        param col_names: list of column names in the order of measurement
        """
        
        data.columns = ['Protein.Group', 'Protein.Ids', 'Protein.Names', 'Genes','First.Protein.Description'] + col_names
        self.col_names = data.columns.unique()
        self.data = data.copy()
        print()
        print('Data has been added - Column names have been changed.')
        print()
        
    def pre_preprocessing_info(self):
        
        # plot subplot of a heatmap and lineplot
        fig, axes = plt.subplots(2,2)
        fig.tight_layout(pad=2)
        fig.set_figheight(8)
        fig.set_figwidth(16)
        
        # plot missing values in lineplot
        x = [str(i) for i in range(len(self.data.iloc[:,5:].columns))]
        y = self.data.iloc[:,5:].isna().sum(axis=0).values
        sns.lineplot(
            y=y,
            x=x,
            marker='o',
            ax=axes[0,0]
        )
        
        # plot missing values in heatmap
        sns.heatmap(
            data=self.data.select_dtypes('float').isnull(),
            cmap='viridis',
            ax=axes[0,1],
            cbar=False,
            xticklabels=True,
            yticklabels=False
        )
        
        # plot number of IDs in barplot
        x = self.data.select_dtypes('float').notna().sum(axis=0).index
        y = self.data.select_dtypes('float').notna().sum(axis=0).values
        sns.barplot(
            x,
            y,
            ci='sd',
            capsize=.4,
            ax=axes[1,0]
        )
        
        # plot the coefficient of variation
        cv_data = pd.DataFrame()

        for i in self.data.select_dtypes('float').columns.unique():
            if type(self.data[i]) != pd.core.series.Series:
                cv_data[i] = (self.data[i].std(axis=1)/data[i].mean(axis=1))*100

        # plot CV within each strain
        cv_data_melt = cv_data.melt()
        sns.violinplot(
            data=cv_data_melt,
            x='variable',
            y='value',
            ax=axes[1,1]
        )
        
        # label the plot
        axes[0,0].set(xlabel='measurment_order')
        axes[0,0].set(ylabel='missing_vals')
        axes[0,1].set(xlabel='measurment_order')
        axes[0,1].set(ylabel='missing_vals')
        axes[1,0].set(xlabel='Sample')
        axes[1,0].set(ylabel='Number_of_ids')
        axes[1,1].set(xlabel='Sample')
        axes[1,1].set(ylabel='CV in %')
        sns.despine()
        plt.show()
    
    def drop_col(self, col_name):
        self.data = self.data.drop(col_name, axis=1)
        return self.data.columns
        
    def preprocessing(self):
        self.data = self._Preprocessor._process(self.data)
        print()
        print('Data has been pre-processed')
        print()
        
```
